```python
def cont_plots(data, col, target):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
    sns.distplot(data[col], ax=axes[0],kde_kws={'bw':0.15}).set_title("Distribution plot")
    sns.boxplot(x=target, y=col, data=data).set_title("Box Plot distribution")
    plt.show()
```

```python
for cl in num_var_list:
    cont_plots(emp_attrition, cl, "Attrition")
```

```python
#Categorical Variable EDA

def cat_plots(idat, col, target = "Attrition"):
    print("Countplot for ",col, " split by ", target)
    print("\n")
    plt.figure(figsize=(20, 10))
    sns.countplot(x=col, data=idat, hue=target)
    plt.show()
```

```python
## Data Preparation 
## Outlier Treatement for numeric data

emp_attrition_data_2 = emp_attrition.copy()

for col in ['YearsWithCurrManager','YearsSinceLastPromotion','YearsInCurrentRole','YearsAtCompany','TrainingTimesLastYear','TotalWorkingYears','StockOptionLevel']:
    sorted(emp_attrition_data_2)
    p1,p99 = np.percentile(emp_attrition_data_2[col],[1, 99])
    lower_bounds = p1
    upper_bounds = p99
    
    emp_attrition_data_2[col] = np.where(emp_attrition_data_2[col] < lower_bounds, lower_bounds,  emp_attrition_data_2[col])
    emp_attrition_data_2[col] = np.where(emp_attrition_data_2[col] > upper_bounds, upper_bounds,  emp_attrition_data_2[col])
```

```python
emp_attrition_data_2.drop(['EmployeeNumber'],axis=1)
emp_attrition_data_2.info()

#Apply One Hot Encodings to the followings columns BusinessTravel,  Department , EducationField , JobRole , Martial Status

BusinessTravel_dummies = pd.get_dummies(emp_attrition_data_2['BusinessTravel'],prefix='BusinessTravel_dummies',drop_first=False)
emp_attrition_data_2 = pd.concat([emp_attrition_data_2,BusinessTravel_dummies],axis=1)
emp_attrition_data_2.drop('BusinessTravel',axis=1,inplace=True)

Department_dummies = pd.get_dummies(emp_attrition_data_2['Department'],prefix='Department_dummies',drop_first=False)
emp_attrition_data_2 = pd.concat([emp_attrition_data_2,Department_dummies],axis=1)
emp_attrition_data_2.drop('Department',axis=1,inplace=True)

#Education_dummies = pd.get_dummies(emp_attrition_data_2['Education'],prefix='Education_dummies',drop_first=False)
#emp_attrition_data_2 = pd.concat([emp_attrition_data_2,Education_dummies],axis=1)
#emp_attrition_data_2.drop('Education',axis=1,inplace=True)

EducationField_dummies = pd.get_dummies(emp_attrition_data_2['EducationField'],prefix='EducationField_dummies',drop_first=False)
emp_attrition_data_2 = pd.concat([emp_attrition_data_2,EducationField_dummies],axis=1)
emp_attrition_data_2.drop('EducationField',axis=1,inplace=True)

#EnvironmentSatisfaction_dummies = pd.get_dummies(emp_attrition_data_2['EnvironmentSatisfaction'],prefix='EnvironmentSatisfaction_dummies',drop_first=False)
#emp_attrition_data_2 = pd.concat([emp_attrition_data_2,EnvironmentSatisfaction_dummies],axis=1)
#emp_attrition_data_2.drop('EnvironmentSatisfaction',axis=1,inplace=True)

#JobInvolvement_dummies = pd.get_dummies(emp_attrition_data_2['JobInvolvement'],prefix='JobInvolvement_dummies',drop_first=False)
#emp_attrition_data_2 = pd.concat([emp_attrition_data_2,JobInvolvement_dummies],axis=1)
#emp_attrition_data_2.drop('JobInvolvement',axis=1,inplace=True)

JobRole_dummies = pd.get_dummies(emp_attrition_data_2['JobRole'],prefix='JobRole_dummies',drop_first=False)
emp_attrition_data_2 = pd.concat([emp_attrition_data_2,JobRole_dummies],axis=1)
emp_attrition_data_2.drop('JobRole',axis=1,inplace=True)

MaritalStatus_dummies = pd.get_dummies(emp_attrition_data_2['MaritalStatus'],prefix='MaritalStatus_dummies',drop_first=False)
emp_attrition_data_2 = pd.concat([emp_attrition_data_2,MaritalStatus_dummies],axis=1)
emp_attrition_data_2.drop('MaritalStatus',axis=1,inplace=True)
```

```python
emp_attrition_data_2.info()
```

```python
# Label Encoding to Gender, OverTime , Over18. Convert data to 0 & 1  

from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
emp_attrition_data_2['Gender']= label_encoder.fit_transform(emp_attrition_data_2['Gender'])
emp_attrition_data_2['OverTime']= label_encoder.fit_transform(emp_attrition_data_2['OverTime'])
emp_attrition_data_2['Over18']= label_encoder.fit_transform(emp_attrition_data_2['Over18'])

# Drop EmployeeNumber, Employee Count, StandardHours. These are stadard values and are not contributing. 

emp_attrition_data_2.drop(['EmployeeNumber','EmployeeCount','StandardHours'],axis=1)
```

```python
## Modelling Data Preparation. 
x = emp_attrition_data_2.drop(["Attrition"],axis =1)
y = emp_attrition_data_2["Attrition"]

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=80, stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=80)
clf.fit(x_train,y_train)
```

```python
pd.Series(clf.feature_importances_ , index= x.columns).sort_values(ascending=False).head(10)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score

## Accuracy Score
accuracy_score(y_test,y_pred)
```

```python
## Precision Score
precision_score(y_test,y_pred)
```

```python
## Recall
recall_score(y_test,y_pred)
```

```python
## F1 Score
f1_score(y_test,y_pred)
```

```python
##Using Grid SearchCV

from sklearn.model_selection import GridSearchCV
param_dict = {
    "criterion":['gini','entropy'],
    "max_depth":range(1,10),
    "min_samples_split":range(1,10),
    "min_samples_leaf":range(1,5)
}

decision_tree = DecisionTreeClassifier(random_state=100)
grid = GridSearchCV(decision_tree,
                   param_grid=param_dict,
                   cv=10,verbose=1,
                   n_jobs=-1)

grid.fit(x_train,y_train)
```

```python
#Plot ROC curve
from sklearn.metrics import roc_curve,auc

def compute_auc(test,train_instance):
    probs = train_instance.predict_proba(test)
    prob_positive = probs[:,1]
    fpr,tpr,threshold = roc_curve(y_test,prob_positive)
    roc_auc = auc(fpr,tpr)
    return roc_auc,fpr,tpr
```

```python
#Compute the area under the curve between fpr and tpr
roc_auc, fpr, tpr = compute_auc(x_test,clf)
roc_auc
```

```python
def plot_ROC(roc,fpr,tpr):
    plt.title("Reciever Operating Characteristics")
    plt.plot(fpr,tpr,'red',label= 'AUC=%0.2f'% roc)
    plt.legend(loc = 'lower right')
    plt.plot([0,1],[0,1], color = 'darkblue',linestyle='--')
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
```

```python
plot_ROC(roc_auc,fpr,tpr)
```
