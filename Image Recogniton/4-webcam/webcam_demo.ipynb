```python

#import keras
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import load_model
import os
import numpy as np

%matplotlib inline

import cv2
import matplotlib.pyplot as plt
from IPython import display

from io import BytesIO
from PIL import Image
from PIL import ImageOps
```

```python
# cifar10 category label name
cifar10_labels = np.array([
    'airplane',
    'automobile',
    'bird',
    'cat',
    'deer',
    'dog',
    'frog',
    'horse',
    'ship',
    'truck'])
```

```python
#transform to a format 
#the model can understand
def convertCIFER10Data(image):
    img = image.astype('float32')
    img /= 255
    c = np.zeros(32*32*3).reshape((1,32,32,3))
    
    c[0] = img
    return c
```

```python
'''
LOAD MODEL
'''

# use ResNet
model = load_model('cifar10_ResNet20v1_model.092.h5')
```

```python
webcam = cv2.VideoCapture(0) #Use camera 0


while True:
    (rval, im) = webcam.read()
    
            
    # Show the image
    cv2.imshow('LIVE',   im)
    key = cv2.waitKey(10)
    # if Esc key is press then break out of the loop 
    if key == 27: #The Esc key
        break

    frame = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)       
        
        # Stop video
webcam.release()

# Close all started windows
cv2.destroyAllWindows()
```

```python
image1 = Image.fromarray(np.uint8(frame))
plt.imshow(image1)
plt.show()
```

```python
# resize cifar10 image size(32, 32)
image = Image.fromarray(np.uint8(frame))
image = image.resize((32, 32))

resize_frame = np.asarray(image)
```

```python
'''
PREDICTING
'''

data = convertCIFER10Data(resize_frame)
 
ret = model.predict(data, batch_size=1) 
 
```

```python
result=ret.argmax()
result
```

```python
label=cifar10_labels[ret.argmax()]
```
